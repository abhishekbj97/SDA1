```python
#import dependencies

import pandas as pd

import numpy as np

import plotly.express as px
```

```python
#load the data into a dataframe

df = pd.read_csv('global_emissions.csv')

#check the first 5 rows

df.head()
```

```python
fig = px.histogram(df, x='Emissions.Production.CH4')

fig.show()
```

```python
#create a box plot

fig = px.box(df, y='Emissions.Production.CH4')

fig.show()
```

```python
fig = px.scatter(x=df['Year'], y=df['Emissions.Production.CH4'])

fig.show()
```

```python
def find_outliers_IQR(df):

    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    return outliers
```

```python
outliers = find_outliers_IQR(df['Emissions.Production.CH4'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

outliers
```

```python
outliers = find_outliers_IQR(df[['Year','Emissions.Production.CH4']])

outliers
```

```python
fig = px.scatter(x=outliers['Year'], y=outliers['Emissions.Production.CH4'])

fig.show()
```

```python
def drop_outliers_IQR(df):

    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    outliers_dropped = not_outliers.dropna().reset_index()
 
    return outliers_dropped
```

```python
drop_outliers_IQR
```

```python
outliers = drop_outliers_IQR(df[['Year','Emissions.Production.CH4']])

outliers
```

```python
fig = px.box(outliers, y='Emissions.Production.CH4')

fig.show()
```
