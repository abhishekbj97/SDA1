```python
import pandas as pd

# Load dataset
df = pd.read_csv('global_emissions.csv')

# Calculate the IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

# Create new dataframe
new_df = pd.DataFrame(df)
```

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset("tips")

# Plot boxplot
sns.boxplot(x="day", y="total_bill", data=df)

# Identify and remove outliers
Q1 = df["total_bill"].quantile(0.25)
Q3 = df["total_bill"].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
df_clean = df[(df["total_bill"] > lower_limit) & (df["total_bill"] < upper_limit)]

# Create new column for outliers
df_clean["outlier"] = ~df_clean["total_bill"].isin(df["total_bill"])

# Print cleaned dataset
print(df_clean.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('global_emissions.csv')
df.head()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['Emissions.Production.CH4'])
plt.show()
# plt.subplot(1,2,2)
# sns.distplot(df['Country.GDP'])
# plt.show()
```

```python
sns.boxplot(df['Emissions.Production.CH4'])
plt.show()
```

```python
percentile25 = df['Emissions.Production.CH4'].quantile(0.25)
percentile75 = df['Emissions.Production.CH4'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
df[df['Emissions.Production.CH4'] > upper_limit]
df[df['Emissions.Production.CH4'] < lower_limit]
```

```python
new_df = df[df['Emissions.Production.CH4'] < upper_limit]
new_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Emissions.Production.CH4'])
plt.subplot(2,2,2)
sns.boxplot(df['Emissions.Production.CH4'])
plt.subplot(2,2,3)
sns.distplot(new_df['Emissions.Production.CH4'])
plt.subplot(2,2,4)
sns.boxplot(new_df['Emissions.Production.CH4'])
plt.show()
```
