```python
Q1 = boston_df.quantile(0.25)
Q3 = boston_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print("Old data set shape with Outliers ",boston_df.shape)
boston_df_out = boston_df[~((boston_df < (Q1 - 1.5 * IQR)) |(boston_df > (Q3 + 1.5 * IQR))).any(axis=1)]
print("New data set shape without Outliers ",boston_df_out.shape)
```

```python
def remove_outlier_Hampel(df):
    med=df.median()
    List=abs(df-med)
    cond=List.median()*4.5
    good_list=List[~(List>cond)]
    return good_list
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("size of the CRIM column before outlier removal is",boston_df['Emissions.Global Share.CO2.Total'].shape[0])
df_outlier_removed=remove_outlier_Hampel(boston_df['Emissions.Global Share.CO2.Total'])
print("size of the CRIM column after outlier removal is",df_outlier_removed.shape[0])
plt.subplot(1,2,1)
sns.boxplot(data=boston_df['Emissions.Global Share.CO2.Total'])
plt.xticks(rotation=90)
plt.ylabel('Price ($)')
# Finding the index's of outlier items 
ind_diff=boston_df.index.difference(df_outlier_removed.index)
# dropping all rows from the data frame which having outliers
for i in range(0, len(ind_diff),1):
    boston_df.drop([ind_diff[i]],inplace=True)
plt.subplot(1,2,2)
sns.boxplot(data=boston_df['Emissions.Global Share.CO2.Total'])
plt.xticks(rotation=90)
```
